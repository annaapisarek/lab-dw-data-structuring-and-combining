In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

Observation:

- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd
import numpy as np
import re
import seaborn as sns

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df_1 = pd.read_csv(url)

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
df_2 = pd.read_csv(url)

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
df_3 = pd.read_csv(url)

df_1.columns = [col.replace(' ', '_').lower() for col in df_1.columns]
df_2.columns = [col.replace(' ', '_').lower() for col in df_2.columns]
df_3.columns = [col.replace(' ', '_').lower() for col in df_3.columns]
df_1.drop_duplicates()
df_2.drop_duplicates()
df_3.drop_duplicates()

new_df = pd.concat([df_1, df_2, df_3], axis=0)

new_df = new_df.drop_duplicates(subset='customer', keep='first')
new_df = new_df.rename({"st":"state"}, axis = 1) 

#Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".
gender_mapping = {
    'F': 'F',
    'f': 'F',
    'Female': 'F',
    'Femal': 'F',
    'M': 'M',
    'm': 'M',
    'Male': 'M',
    'male': 'M'
}

new_df['gender'] = new_df['gender'].replace(gender_mapping)

#State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"
state_mapping = {
    'AZ': 'Arizona',
    'Cali': 'California',
    'WA': 'Washington'
}

new_df['state'] = new_df['state'].replace(state_mapping)

#In education, "Bachelors" could be replaced by "Bachelor"
mapping = {
    'Bachelors': 'Bachelor'}

new_df['education'] = new_df['education'].replace(mapping)

#In Customer Lifetime Value, delete the % character
new_df['customer_lifetime_value'] = new_df['customer_lifetime_value'].str.replace("%", "", regex=False)

#In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"
new_df['vehicle_class'] = new_df['vehicle_class'].str.replace("Sports Car", "Luxury", regex=False)
new_df['vehicle_class'] = new_df['vehicle_class'].str.replace("Luxury SUV", "Luxury", regex=False)
new_df['vehicle_class'] = new_df['vehicle_class'].str.replace("Luxury Car", "Luxury", regex=False)

new_df['customer_lifetime_value'] = new_df['customer_lifetime_value'].astype(float) 




mean_income = new_df["income"].mean()
new_df["income"].fillna(mean_income, inplace=True) 


mean_clv = new_df['customer_lifetime_value'].mean()
new_df['customer_lifetime_value'].fillna(mean_clv, inplace=True) 


df_insurance.head()

```
